```python
def plot_feature_vs_target(df,features,target):
    n=len(features)
    rows=(n+1)//2
    fig,axes=plt.subplots(rows,2,figsize=(14,rows*7))
    for x,i in enumerate(features):
        row,col=divmod(x,2)
        sns.scatterplot(x=df[i],y=df['MEDV'],ax=axes[row,col])
        axes[row,col].set_title(f'{i}vs{'MEDV'}')
    if n % 2!=0:
        fig.delaxes(axes[rows-1,1])
    plt.tight_layout()
    plt.show()
```

```python
plot_feature_vs_target(df,features,target)
```

```python
# Box plots to visulaize the outliers.
plt.figure(figsize=(20,15))
for i, column in enumerate(df.columns,1):
    plt.subplot(3,5,i)
    sns.boxplot(df[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df,columns):
    for column in columns:
        q1=df[column].quantile(0.25)
        q3=df[column].quantile(0.75)
        iqr=q3-q1
        lower_bound=q1-1.5*iqr
        upper_bound=q3+1.5*iqr
        df=df[(df[column]>=lower_bound) & (df[column]<=upper_bound)]
    return df
```

```python
df_cleaned=remove_outliers(df,df.columns)
```

```python
print(f"Shape of the original dataset: {df.shape}")
print(f"Shape of the cleaned dataset: {df_cleaned.shape}")
```
